```python
from bs4 import BeautifulSoup as bs
import urllib.request as urlrequest
```

```python
res_area_url = 'https://sh.lianjia.com/xiaoqu/5011102206884/'
web_page = urlrequest.urlopen(res_area_url).read()
soup = bs(web_page, 'html.parser')
print(soup.prettify())
```

```python
info_label_list = soup.find_all(class_ = 'xiaoquInfoLabel')
info_content_list = soup.find_all(class_ = 'xiaoquInfoContent')
```

```python
from collections import defaultdict
#song_database = defaultdict(list)
#for song in l.songs:
#    song_database[song.album].append(song.name)
```

```python
info_dict = defaultdict(list)
for i,j in zip(info_label_list, info_content_list):
    info_dict[i.get_text()].append(j.get_text())
info_dict
```

```python
import pandas as pd
a = pd.DataFrame(info_dict)
a
```

```python
b = a.drop(a.columns[0], axis=1, inplace=False)
b
```
