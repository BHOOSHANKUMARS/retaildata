```python
import pandas as pd
data = pd.read_csv('./data.csv')
data
```

```python
import folium
resid = folium.map.FeatureGroup()
for i in range(len(data)):
    location = data.iloc[i,6].split(',')
    location = list(map(float, location))
    resid.add_child(
        folium.CircleMarker(location,
                            radius=7,
                            color='yellow',
                            fill=True,
                            fill_color='red',
                            fill_opacity=0.4
                           )
    )
resid_map = folium.Map(location = [121.486477,30.918412],
                       zoom_start = 12
                       )
resid_map.add_child(resid)
```
